# Construct Binary Tree from Preorder and Inorder Traversal

Given two integer arrays preorder and inorder where preorder is the preorder traversal of a binary tree and inorder is the inorder traversal of the same tree, construct and return the binary tree.

 

Example 1:


Input: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]
Output: [3,9,20,null,null,15,7]
Example 2:

Input: preorder = [-1], inorder = [-1]
Output: [-1]

# Solution:

We can traverse a binary tree several ways, lets see first what is inorder and preorder traversal

Inorder traversal: First visit the left node, then root, right at last


Preorder traversal: First visit root, then left node, right node at last 


We can construct the binary tree with the help of inorder and preorder traversal. We can always find the root node using preorder:

1) Pick an element from Preorder. Maintain a Preorder Index Variable to pick the next element in the next recursive call. 

2) Create a treenode, if the picked element is not present in the inorder increase the index

3) If the picked element is in inorder, set it as node value

4) Increase the index, Call dfs for elements before picked element index in inorder and assing it as left subtree of treenode. 

5) Increase the index, Call dfs for elements after picked element index in inorder and assing it as right subtree of treenode. 

6) return treenode.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def buildTree(self, preorder, inorder):
        # build tree using dfs
        def dfs(index,inorder):
            # if inorder array is blank return Null
            if len(inorder)==0: return None
            # initialize tree node
            node=TreeNode()
            # while index is within range or preorder
            while index<len(preorder):
                if preorder[index] in inorder:
                    # if picked value from preorder is present in inorder
                    val=preorder[index]
                    # set it as node value
                    node.val=val
                    # increment the index,call dfs for right and left nodes
                    node.left=dfs(index+1,inorder[:inorder.index(val)])
                    node.right=dfs(index+1,inorder[inorder.index(val)+1::])
                    break
                # increment the index if picked inorder index is not in inorder    
                index+=1   
            # return Node    
            return node
        return dfs(0,inorder)
                    
                    
```
