## Permutation in String


Share
Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

 

Example 1:

Input: s1 = "ab", s2 = "eidbaooo"
Output: true
Explanation: s2 contains one permutation of s1 ("ba").

Example 2:

Input: s1 = "ab", s2 = "eidboaoo"
Output: false
 

Constraints:

1 <= s1.length, s2.length <= 104
s1 and s2 consist of lowercase English letters.

```python
class Solution(object):
    def checkInclusion(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: bool
        """
        # if length s1 is greater than s2 return False
        if len(s1)>len(s2): return False
        # initialize two list to hold letter count for each string(a-z)
        s1Count,s2Count=[0]*26,[0]*26
        # initialize a variable to keep track a number of matches
        matches=0
        # initialize the count trackers upto the index of length of s1
        for i in range(len(s1)):
            s1Count[ord(s1[i])-ord('a')]+=1
            s2Count[ord(s2[i])-ord('a')]+=1
        # calculate the number of matches untill this point    
        for i in range(26):
            if s1Count[i]==s2Count[i]:
                matches+=1
        # initialize a pointer to keep a track at the left of sliding window        
        l=0
        # move the right pointer of the sliding window
        for r in range(len(s1),len(s2)):
            # the the matches is already 26 return True
            if matches==26: 
                return True
            # calculate the index for the right pointer of the sliding window
            # we are adding a new element here in the sliding window
            index=ord(s2[r])-ord('a')
            s2Count[index]+=1
            # if the counts became same after adding the current element, increase match counts 
            if s1Count[index]==s2Count[index]:
                matches+=1
            # if the count was same before, reduce the count of matches    
            elif s1Count[index]+1==s2Count[index]:
                matches-=1
            # calculate the index for the left pointer of the sliding window
            # we are removing a element from the begining of the sliding window
            index=ord(s2[l])-ord('a')
            s2Count[index]-=1
            # if the counts became same after removing the left element, increase match counts 
            if s1Count[index]==s2Count[index]:
                matches+=1
            # if the count was same before removing the left element, reduce the count of matches    
            elif s1Count[index]-1==s2Count[index]:
                matches-=1    
            l+=1    
        if matches==26: return True   
        return False   
            
            
        
```
