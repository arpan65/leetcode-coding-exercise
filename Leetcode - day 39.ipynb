# Binary Tree Right Side View

Given the root of a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

 

Example 1:


Input: root = [1,2,3,null,5,null,4]
Output: [1,3,4]
Example 2:

Input: root = [1,null,3]
Output: [1,3]

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def rightSideView(self, root:
        visited=[]
        def bfs(node):
            if not node:
                return
            queue=[]
            queue.append(node)
            while queue:
                right=None
                for i in range(len(queue)):
                    cur=queue.pop(0)
                    if cur:
                        right=cur
                        queue.append(cur.left)
                        queue.append(cur.right)
                if right:    
                    visited.append(right.val) 
        bfs(root)
        return visited
        
```

# All Nodes Distance K in Binary Tree

Given the root of a binary tree, the value of a target node target, and an integer k, return an array of the values of all nodes that have a distance k from the target node.

You can return the answer in any order.

 

Example 1:


Input: root = [3,5,1,6,2,0,8,null,null,7,4], target = 5, k = 2
Output: [7,4,1]
Explanation: The nodes that are a distance 2 from the target node (with value 5) have values 7, 4, and 1.
Example 2:

Input: root = [1], target = 1, k = 3
Output: []

# Solution:

1. Trace parent node of each node
2. run BFS (level order traversal), trace depth from target node
3. while depth=k return queue

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def distanceK(self, root, target, k) :
        queue=[]
        def dfs(node,parent=None):
            if node:
                node.par=parent
                if node.val==target.val:
                    target.par=parent
                if node.right:
                    dfs(node.right,node)
                if node.left:
                    dfs(node.left,node)
        dfs(root)
        visited=[]
        visited.append(target)
        queue.append((target,0))
        level=0
        while len(queue)>0:
            if queue[0][1]==k:
                return ([ele.val for ele,l in queue])
            cur=queue.pop(0)
            if cur[0].par and (cur[0].par not in visited):
                visited.append(cur[0].par)    
                queue.append((cur[0].par,cur[1]+1))
            if cur[0].left and (cur[0].left not in visited):
                queue.append((cur[0].left,cur[1]+1))    
                visited.append(cur[0].left)
            if cur[0].right  and (cur[0].right not in visited):
                queue.append((cur[0].right,cur[1]+1))  
                visited.append(cur[0].right)
        return []
```
