## Search Insert Position

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2

Example 2:

Input: nums = [1,3,5,6], target = 2
Output: 1

Example 3:

Input: nums = [1,3,5,6], target = 7
Output: 4

```python
class Solution:
    default=-1
    def searchInsert(self, nums, target):
        def search(nums,l,r):
            # define mid of two pointers
            mid=int((l+r)/2)
            if l<=r:
                # if target is the mid element return index
                if nums[mid]==target:
                    return mid
                # if mid element is less than target, mid+1 can be the right place to insert
                elif nums[mid]<target:
                    self.default=mid+1
                    return search(nums,mid+1,r)
                else:
                    # else mid can be the right index to insert the element
                    self.default=mid
                    print(self.default)
                    return search(nums,l,mid-1)
            else:
                # return the index if the element is not present
                return self.default
        return search(nums,0,len(nums)-1)     
                
```

## Squares of a Sorted Array
https://leetcode.com/problems/squares-of-a-sorted-array/

Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.

 

Example 1:

Input: nums = [-4,-1,0,3,10]
Output: [0,1,9,16,100]
Explanation: After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100].

Example 2:

Input: nums = [-7,-3,2,3,11]
Output: [4,9,9,49,121]
 

Constraints:

1 <= nums.length <= 104
-104 <= nums[i] <= 104
nums is sorted in non-decreasing order.


```python
class Solution:
    default=-1
    def sortedSquares(self, nums):
        # def
        l,r=0,len(nums)-1
        res=[0]*len(nums)
        for i in range(len(nums)-1,-1,-1):
            if abs(nums[l])>=abs(nums[r]):
                res[i]=abs(nums[l])**2
                l+=1
            elif abs(nums[l])<abs(nums[r]):
                res[i]=abs(nums[r])**2
                r-=1
        return res        
```

## Rotate Array
https://leetcode.com/problems/rotate-array/

Example 1:

Input: nums = [1,2,3,4,5,6,7], k = 3
Output: [5,6,7,1,2,3,4]

Explanation:
rotate 1 steps to the right: [7,1,2,3,4,5,6]
rotate 2 steps to the right: [6,7,1,2,3,4,5]
rotate 3 steps to the right: [5,6,7,1,2,3,4]

Example 2:

Input: nums = [-1,-100,3,99], k = 2
Output: [3,99,-1,-100]

Explanation: 
rotate 1 steps to the right: [99,-1,-100,3]
rotate 2 steps to the right: [3,99,-1,-100]
 

Constraints:

1 <= nums.length <= 105
-231 <= nums[i] <= 231 - 1
0 <= k <= 105
 

Follow up:

Could you do it in-place with O(1) extra space?

```python
class Solution:
    def rotate(self, nums, k) :
        """
        Do not return anything, modify nums in-place instead.
        """
        l,r=0,len(nums)-1
        while l<r:
                nums[l],nums[r]=nums[r],nums[l]
                l+=1
                r-=1 
        k=k % len(nums)
        l,r=0,k-1
        while l<r:
                nums[l],nums[r]=nums[r],nums[l]
                l+=1
                r-=1 
        l,r=k,len(nums)-1        
        while l<r:
                nums[l],nums[r]=nums[r],nums[l]
                l+=1
                r-=1 
            
```
