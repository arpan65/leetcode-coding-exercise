# Minimum Depth of Binary Tree

Given a binary tree, find its minimum depth.

The minimum depth is the number of nodes along the shortest path from the root node down to the nearest leaf node.

Note: A leaf is a node with no children.

 

Example 1:


Input: root = [3,9,20,null,null,15,7]
Output: 2


Example 2:

Input: root = [2,null,3,null,4,null,5,null,6]
Output: 5

## Solution:

The problem can be solved using BFS. We will traverse the graph level wise and check if there is a leaf node at that level. Once we found a leaf node at any level, we will stop computation and return the current level as minimum depth.

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def minDepth(self, root: [TreeNode]):
        # if root is empty itself return level=0
        if root is None:
            return 0
        # initialize level, queue and temp
        level=1
        queue,temp=[],[]
        queue.append(root)
        # while queue is not empty
        while queue:
            # pop an element from the queue
            cur=queue.pop(0)
            # if the current node is a child node, return the current level as min depth
            if (cur is not None and cur.left is None and cur.right is None):
                return level
            # otherwise if cur node has a left child, add it to the temp
            if (cur is not None and cur.left is not None):
                temp.append(cur.left)
            # similarly if cur node has a right child, add it to the temp    
            if (cur is not None and cur.right is not None):
                temp.append(cur.right)   
            # when the current level is exhusted but temp is not empty    
            if len(queue)==0 and len(temp)>0:
                # set queue with temp value, reinitialize temp for next level, increse level counter
                queue=temp
                temp=[]
                level+=1
        return level        
        
```
