## Insertion Sort

Insertion sort is an inplace sorting algorithm, where at each iteration we take the first element from the unsorted array and <b>insert</b> it in to the proper place of the sorted part of the array.

inplace sorting - needs constant number of variables, without any extra space/ data structure 

online sorting - https://www.youtube.com/watch?v=RT_4x38MXAU&t=8s

```python
def insertionSort(arr):
    # from 2nd element to the rest of the input array
    for j in range(1,len(arr)):
        # key = array[j]
        key=arr[j]
        # i=j-1
        i=j-1
        # while i greater than equal to 0 and the array[i] is greater than the key
        while i>=0 and arr[i]>key:
            # keep swapping the elements and decrement the counter
            arr[i+1]=arr[i]
            i-=1
        # replace the i+1 th element with the key    
        arr[i+1]=key
    return arr
        
```

```python
insertionSort([33,55,6,78,9,0,44,555,6,78,87,2323,556,67,89])
```
