## Merge Two Sorted Lists

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:


Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]

Example 2:

Input: list1 = [], list2 = []
Output: []

Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
 

Constraints:

The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # initialize a dummy node, init is used to keep track of the starting pointer
        cur=init=ListNode()
        # while both list1 and list2 have nodes to evaluate
        while list1 and list2:
            # if node at list1 is smaller or equal to node at l2, add the l1 node as cur next
            # increment the list1 pointer to point the next node
            if list1.val<=list2.val:
                cur.next=list1
                list1=list1.next
            # if node at list2 is smaller or equal to node at l1, add the l2 node as cur next
            # increment the list2 pointer to point the next node    
            else:
                cur.next=list2
                list2=list2.next
            # increment the cur pointer to point the next node    
            cur=cur.next
        # if list2 is empty and only list1 is left, add list1 as the next of cur node
        if list1:
            cur.next=list1
        # otherwise if list1 is empty and only list2 is left, add list2 as the next of cur node    
        elif list2:
            cur.next=list2
        return init.next
        
```

## Reverse Linked List

Example 1:


Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
Example 2:


Input: head = [1,2]
Output: [2,1]
Example 3:

Input: head = []
Output: []
 

Constraints:

The number of nodes in the list is the range [0, 5000].
-5000 <= Node.val <= 5000

### Approach1: Iterative

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # declaire two pointers current nad previous
        prev,cur=None,head
        # while cur is not none
        while cur:
            # save current next pointer
            next=cur.next
            # change next pointer of current node to previous
            cur.next=prev
            # point current pointer as previous node for the next node
            prev=cur
            # go to the next node of the original list 
            cur=next
        # current is now none, return the prev pointer as answer
        return prev    
            
           
```

### Approach 2 : Recursive

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # base case : if the head is null, return None
        if not head:
            return None
        # initialize a pointer current head
        curHead=head
        # recursively traserve the list untill the last node
        while head.next:
            # update current head
            curHead=self.reverseList(head.next)
            # update the next pointer
            head.next.next=head
            head.next=None
        head.next=None 
        return curHead
            
           
```
